```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('ecommerce_preparados.csv')


plt.hist(df['Qtd_Vendidos_Cod'])
plt.title('Histograma: Qtd Vendidos')
plt.xlabel('Qtd de Vendas')
plt.ylabel('Frequência')
plt.show()

sns.scatterplot(y='Desconto', x='Qtd_Vendidos_Cod', data=df)
plt.title('Gráfico de Dispersão: Qtd Vendidos vs. Desconto')
plt.xlabel('Desconto')
plt.ylabel('Qtd de Vendas')
plt.show()


df_corr = df[['Material_Cod', 'Qtd_Vendidos_Cod', 'Nota', 'Desconto']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(df_corr, annot=True, fmt=".2f")
plt.title('Mapa de Calor: correlação entre variáveis')
plt.show()


plt.bar(df['Nota'], df['Qtd_Vendidos_Cod'])
plt.title('Qtd de Vendas por nota')
plt.xlabel('Nota')
plt.ylabel('Qtd de Vendas')
plt.xticks(rotation=90)
plt.show()

df['Nota_Agrupada'] = df['Nota'].apply(lambda x: 'SEM NOTA' if pd.isnull(x) else ('> 4' if x < 4.0 else str(x)))

x = df['Nota_Agrupada'].value_counts().index
y = df['Nota_Agrupada'].value_counts().values

plt.figure(figsize=(10, 6))
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=90)
plt.title('Porcentagem de Avaliações')
plt.xlabel('Nota')
plt.ylabel('Quantidade')
plt.show()

plt.figure(figsize=(10, 6))
sns.kdeplot(df['Qtd_Vendidos_Cod'])
plt.title('Densidade de Vendas')
plt.xlabel('Qtd Vendidos')
plt.ylabel('Densidade')
plt.show()


sns.regplot(x='Desconto', y='Qtd_Vendidos_Cod', data=df , color='#278f65' , scatter_kws={'alpha':0.5, 'color': '#34c289'})
plt.title('Gráfico de Regressão: Qtd Vendidos vs. Nota')
plt.xlabel('Desconto')
plt.ylabel('Qtd de Vendas')
plt.show
```
